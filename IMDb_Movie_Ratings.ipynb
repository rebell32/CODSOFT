```python
#dropping columns which are not required
df_new= df[['Genre','Director','Actor 1','Actor 2','Actor 3','Rating']]
df_new.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df_new['Genre']=label_encoder.fit_transform(df_new['Genre'])
df_new['Director']=label_encoder.fit_transform(df_new['Director'])
df_new['Actor 1']=label_encoder.fit_transform(df_new['Actor 1'])
df_new['Actor 2']=label_encoder.fit_transform(df_new['Actor 2'])
df_new['Actor 3']=label_encoder.fit_transform(df_new['Actor 3'])

print(df_new.head())
```

```python
df_new.head()
```

```python
df_new.isnull().sum()
```

```python
df_new= df_new.copy()
df_new.dropna(subset=['Genre','Director','Actor 1','Actor 2','Actor 3','Rating'],inplace=True)
```

```python
df_new.head()
```

```python
import plotly.express as px

fig = px.scatter(df_new, x='Director', y='Rating',hover_data=['Genre','Actor 1','Actor 2','Actor 3'], title ='Movie Ratings')
fig.show()
```

```python
from bokeh.plotting import figure,output_notebook, show
from bokeh.models import HoverTool
from bokeh.io import push_notebook

output_notebook()

p=figure(title ='Movie Ratings', x_axis_label='Director',y_axis_label='Rating')

p.circle('Director', 'Rating', source=df_new)

hover= HoverTool(tooltips=[('Genre','@Genre'), ('Actor 1', '@{Actor 1}'), ('Actor 2', '@{Actor 2}'), ('Actor 3', '@{Actor 3}')])
p.add_tools(hover)

show(p)
```
